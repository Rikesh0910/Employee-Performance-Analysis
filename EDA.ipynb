```python
top_df = df.groupby("Name")["Productivity (%)"].agg('max').sort_values(ascending=False).head(10).reset_index()
```

```python
df_gender = df.groupby("Gender")["Salary"].agg("mean").reset_index()

df_gender
```

```python
df_dept_satisfaction = df.groupby("Department")["Satisfaction Rate (%)"].agg("mean").sort_values(ascending=False).reset_index()

df_dept_satisfaction
```

```python
plt.Figure(figsize=(10, 6))
sns.barplot(y="Satisfaction Rate (%)", x="Department", data=df_dept_satisfaction, palette="viridis")
plt.title("Employee Average Satisfaction Rate by Department")
plt.ylabel("Satisfaction Rate")
plt.xlabel("Department")
plt.show()
```

```python
df_age_satisfaction = df.groupby("Age")["Satisfaction Rate (%)"].agg('mean').sort_values(ascending=False).reset_index()

df_age_satisfaction
```

```python
plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")
sns.regplot(
    x='Age',
    y='Satisfaction Rate (%)',
    data=df_age_satisfaction,
    scatter=True,     
    color='steelblue',
    line_kws={'color': 'darkred', 'linewidth': 2}  
)
plt.title("Employee Satisfaction Rate by Age")
plt.xlabel("Age")
plt.ylabel("Satisfaction Rate")
plt.show()
```

# Insights:

- There's a slight upward trend, suggesting older employees may be slightly more satisfied, though not drastically.

- The confidence interval (shaded area) shows that the prediction is fairly uncertain in some age ranges, likely due to data spread or outliers

```python
df_gender_satisfaction = df.groupby("Gender")["Satisfaction Rate (%)"].agg('mean')

df_gender_satisfaction
```

```python
plt.figure(figsize=(8,6))
plt.pie(df_gender_satisfaction.values,
    labels=df_gender_satisfaction.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=['#66b3ff','#ff9999'],  # Optional color customization
    explode=[0.05, 0] )
plt.title("Employee Satisfaction Rate by Gender")
plt.show()
```

## Insights

```python
df_proj_comp_emp = df.groupby("Name")["Projects Completed"].agg("max").sort_values(ascending=False).head(10).reset_index()

df_proj_comp_emp
```

```python
def performance_level(row):
    if row['Productivity (%)'] >= 80 and row['Satisfaction Rate (%)'] >= 70:
        return 'High Performer'
    elif row['Productivity (%)'] < 50:
        return 'Low Performer'
    else:
        return 'Moderate'

df['Performance Level'] = df.apply(performance_level, axis=1)
```

```python
df_emp_performance = df['Performance Level'].value_counts()
```

```python
df_high_performers = df[df["Performance Level"] == "High Performer"]

def high_performers_age_seg(row):
    if row["Age"] >= 20 and row["Age"] <= 29:
        return "20 - 29"
    elif row["Age"] >= 30 and row["Age"] <= 39:
        return "30 - 39"
    elif row["Age"] >= 40 and row["Age"] <= 49:
        return "40 - 49"
    
df_high_performers["Age Group"] = df_high_performers.apply(high_performers_age_seg, axis=1)
```

```python
df_high_performers
```

```python
df_final_hp_age_group = df_high_performers["Age Group"].value_counts().reset_index()
```

```python
plt.pie(
    df_final_hp_age_group['count'],
    labels=df_final_hp_age_group['Age Group'],
    autopct='%1.1f%%',
    startangle=140,
    colors=sns.color_palette('pastel')
)
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title("High Performers Age Distribution")
plt.show()
```

# Insights:

- The 20â€“29 age group accounts for the majority of high performers (58.3%).

- This suggests that younger employees are currently more engaged, productive, or better aligned with performance metrics in our organization.

- The younger talent pool is thriving â€” they may benefit from early leadership development, skill enhancement, or career acceleration tracks.
